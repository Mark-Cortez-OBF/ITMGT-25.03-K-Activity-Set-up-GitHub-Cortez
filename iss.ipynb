```python
import requests
import json
import time

timestampdata=[]
longitudedata=[]
latitudedata=[]
count=0

while count!=100:
    req = requests.get("http://api.open-notify.org/iss-now.json")
    obj = json.loads(req.text)
    timestampdata.append(float(obj['timestamp']))
    longitudedata.append(float(obj['iss_position']['longitude']))
    latitudedata.append(float(obj['iss_position']['latitude']))
    count=count+1
    time.sleep(10)
    
print(timestampdata)
print(longitudedata)
print(latitudedata)
```
